```python
def date_feature(df):
    df["year"] = df["date"].apply(lambda x: str(x)[:4])
    df["month"] = df["date"].apply(lambda x: str(x)[4:6])
    df["day"] = df["date"].apply(lambda x: str(x)[6:8])
    df["date"] = pd.to_datetime(df[["year", "month", "day"]])
    return df
```

```python
option_2015_1 = pd.read_csv("OptionPrices_2015_m1_m6.csv")
option_2015_2 = pd.read_csv("OptionPrices_2015_m7_m12.csv")
option_2016 = pd.read_csv("OptionPrices_2016.csv")

option_2015_1 = date_feature(option_2015_1)
option_2015_2 = date_feature(option_2015_2)
option_2016 = date_feature(option_2016)
```
